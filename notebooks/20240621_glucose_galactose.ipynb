```python
def invert_dict(input_dict):
    """Invert a dictionary ignore None values."""
    # Create an empty dictionary to store the inverted mapping
    output_dict = {}

    # Iterate over items in the input dictionary
    for key, value in input_dict.items():
        if value is not None:
            if value in output_dict:
                output_dict[value].append(key)
            else:
                output_dict[value] = [key]

    return output_dict
```

```python
names = ['Time', 'T'] + ['{}{}'.format(x, y) for x in ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'] for y in np.arange(1, 13)]
df = pd.read_csv("../data/20240620_galactose_glucose_diauxie.txt", names=names).dropna()

def parse_time(x):
    arr = x.split(':')
    time = int(arr[1]) + 60 * int(arr[0]) + int(arr[2]) / 60
    return time
    
df['time [min]'] = df['Time'].apply(parse_time)
df.drop(columns=['Time', 'T'], inplace=True)

df = pd.melt(df, id_vars=['time [min]'], value_vars=df.columns[:-1], var_name='well', value_name='OD600')
df = df.loc[df['well'] != 'A12', :]
df = df.loc[df['well'] != 'G12', :]
df = df.loc[df['well'] != 'H12', :]
df['replicate'] = df['well'].apply(lambda x: well_metadata[x].split('_')[1] if well_metadata[x]!= 'blank' else 1)
df['cond'] = df['well'].apply(lambda x: well_metadata[x].split('_')[0])
df.head(5)
```

```python
df_merge = pd.merge(df_mean, df_blanks, on="time [min]")
df_merge['OD600'] = df_merge['OD600_x'] - df_merge['OD600_y']
df_merge.drop(columns=['OD600_x', 'OD600_y'], inplace=True)
df_merge
```

```python
# Define exponential growth function
def exp_growth(t, A0, r):
    return A0 * np.exp(r * t)

def fit_growth(t, A):
    """Obtain estimates for A0 and r for a given sub-trace"""
    # Parameter guesses
    A0_guess = 0.01
    r_guess = 1/60  # 1/hr

    # Pull out data


    # Perform optimization
    popt, _ = scipy.optimize.curve_fit(exp_growth, t, A, p0=(A0_guess, r_guess))

    return popt

times = {   
    "1:0": [200, 500],
    "0:1": [400, 800],
    "1:100.0": [400, 600],
    "1:31.62": [400, 600],
    "1:56.23": [400, 600],
    "1:17.78": [500, 700],
    "1:10.0": [600, 650],
    "1:5.62": [620, 650],
}
results = []
for cond in list(times.keys()):
    for rep in ['1', '2']:
        t_start, t_end = times[cond]
        sub_df = df_merge.loc[(df_merge["time [min]"] >= t_start) & \
                              (df_merge["time [min]"] <= t_end) & \
                              (df_merge['cond'] == cond) & \
                              (df_merge['replicate'] == rep), :]
        
        t = sub_df["time [min]"].values
        A = sub_df["OD600"].values
        A0, r = fit_growth(t, A)
        results.append(dict(cond=cond, replicate=rep, A0=A0, r=r))

# Make a data frame from the list of results
df_growth_rates = pd.DataFrame(results)

# Take a look
df_growth_rates
```

```python
order = (
    ("1:0", '1'),
    ("1:0", '2'),
    ("1:5.62", '1'),
    ("1:5.62", '2'),
    ("1:10.0", '1'),
    ("1:10.0", '2'),
    ("1:17.78", '1'),
    ("1:17.78", '2'),
    ("1:31.62", '1'),
    ("1:31.62", '2'),
    ("1:56.23", '1'),
    ("1:56.23", '2'),
    ("1:100.0", '1'),
    ("1:100.0", '2'),
    ("0:1", '1'),
    ("0:1", '2'),
)

# Colors
colors = ['#1f77b4', 'orange'] * 8


p = iqplot.strip(
    df_growth_rates,
    q="r",
    cats=['cond', 'replicate'],
    q_axis='y',
    y_axis_label="growth rate (1/min)",
    spread="swarm",
    frame_width=500,
    frame_height=300,
    order=order,
    palette=colors,
    x_axis_label="Glucose:Galactose"
)

bokeh.io.show(p)
```
