```python
def invert_dict(input_dict):
    """Invert a dictionary ignore None values."""
    # Create an empty dictionary to store the inverted mapping
    output_dict = {}

    # Iterate over items in the input dictionary
    for key, value in input_dict.items():
        if value is not None:
            if value in output_dict:
                output_dict[value].append(key)
            else:
                output_dict[value] = [key]

    return output_dict
```

```python
conditions = invert_dict(well_metadata)

conditions
```

```python
df_list = []
with open("../data/test.txt") as f:
    msg = f.read().strip().split("Read")
    for item in msg[1:]:
        df = pd.read_table(StringIO("\n".join(item.split('\n')[2:])), sep='\t')
        df_list.append(df)
```

```python
df_list[0].values[0].split(':')
```

```python
def parse_time(x):
    arr = x.split(':')
    time = int(arr[1]) + 60 * int(arr[0]) + int(arr[2]) / 60
    return time
```

```python
for df in df_list:
    df['time [min]'] = df['Time'].apply(parse_time)

for i in range(1, len(df_list)):
    df_list[i]['time [min]'] += df_list[i-1]['time [min]'].values[-1]

df = pd.concat(df_list).reset_index(drop=True)
df.drop(columns=['Time', 'T '], inplace=True)
```

```python
# Background traces
p = bokeh.plotting.figure(
    frame_width=400,
    frame_height=200,
    x_axis_label='time [min]',
    y_axis_label='absorbance',
)

# Plot all background traces
for well in conditions['blank']:
    p.line(x=df['time [min]'], y=df[well])

# Compute background by averaging background values at each time point
df['background'] = df[conditions['blank']].mean(axis=1)

# Overlay average blank
p.line(x=df['time [min]'], y=df['background'], color='tomato', line_width=2)

bokeh.io.show(p)
```

Spikes are visible in the OD. These are at time points where the plate reader opened so I can add Glucose to the next well.

```python
# Instantiate figure
p = bokeh.plotting.figure(
    frame_width=400,
    frame_height=200,
    x_axis_label='time [min]',
    y_axis_label='absorbance',
)

# Loop through and add a line for each pure glucose trace
for well in conditions['Galactose 0h']:
    p.line(x=df['time [min]'], y=df[well] - df['background'])

bokeh.io.show(p)
```

```python
# Instantiate figure
p = bokeh.plotting.figure(
    frame_width=700,
    frame_height=400,
    x_axis_label='time [min]',
    y_axis_label='absorbance',
    y_axis_type='log',
    x_range=[0, 700],
)

# Loop through and add a line for each pure glucose trace
legend_items = []
# Loop through conditions with pretty colors
for condition, color in zip(['Lactose 0h', 'Lactose 1h', 'Lactose 2h', 'Lactose 3h', 'Lactose 4h', 'Lactose 5h', 'Lactose 6h'], colorcet.b_glasbey_category10):
    line = p.line(x=df['time [min]'], y=df[conditions[condition]].mean(axis=1) - df['background'], color=color)
    legend_items.append((condition, [line]))



# Create the legend from the items
legend = bokeh.models.Legend(items=legend_items, click_policy="hide")

# Add the legend to the plot
p.add_layout(legend, "right")

bokeh.io.show(p)
```
